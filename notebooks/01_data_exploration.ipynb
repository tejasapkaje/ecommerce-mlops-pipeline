# Phase 1: Data Ingestion & Preprocessing

```python
# Cell 1: Importing libraries
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

# Load the dataset
# Are on a 1 folder (notebooks), so '../data/' use this path to access the data folder which is at the same level as notebooks
file_path = '../data/clothing_dataset_200k.csv'
df = pd.read_csv(file_path)

print("Data loaded successfully!")
print(f"Dataset Shape: {df.shape}")
```

```python
# Cell 2: Check missing values & data types
print("--- Missing Values ---")
print(df.isnull().sum())

print("\n--- Data Info ---")
df.info()
```

```python
# Cell 3: Handling Missing Values

# 1. Age: Fill missing ages with the median age (approx 41)
df['age'] = df['age'].fillna(df['age'].median())

# 2. Discount: If discount is missing, then assume 0 discount was given
df['discount'] = df['discount'].fillna(0)

# 3. Rating: Fill missing ratings with the median rating (3.0)
df['rating'] = df['rating'].fillna(df['rating'].median())

# 4. City: Fill missing cities with the most common city (Mode)
most_common_city = df['city'].mode()[0]
df['city'] = df['city'].fillna(most_common_city)

# Check again to ensure no missing values remain
print("Missing values after cleaning:")
print(df.isnull().sum())
```

```python
# Cell 4: Save cleaned dataset
cleaned_path = '../data/cleaned_clothing_dataset.csv'
df.to_csv(cleaned_path, index=False)
print(f"Cleaned data saved to {cleaned_path}")
```
